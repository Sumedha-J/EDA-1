```python
terrorism["Target_Type"].value_counts().plot(kind = "bar", color = "red", figsize=(16,4))
plt.title("Distribution of Target Type")
plt.xlabel("Target Type")
plt.xticks(fontsize = 12)
plt.ylabel("Number of Attacks")
plt.show()
```

```python
terrorism["Gang_Name"].value_counts()[1:16].plot(kind = "bar", color = "blue", figsize=(10,6))
plt.title("Distribution of Gang wise Attack")
plt.xlabel("Gang Name")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
terrorism["City"].value_counts()[1:16].plot(kind = "bar", color = "yellow", figsize=(10,6))
plt.title("Distribution of killed wise Attack")
plt.xlabel("Number of killed")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
terrorism["Country"].value_counts()[1:16].plot(kind = "bar", color = "yellow", figsize=(10,6))
plt.title("Distribution of killed wise Attack")
plt.xlabel("Number of killed")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
terrorism["Attack_Type"].value_counts()[1:16].plot(kind = "bar", color = "yellow", figsize=(10,6))
plt.title("Distribution of Attack_Type wise Attack")
plt.xlabel("Number of Attack_Type")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
pd.crosstab(terrorism.Year,terrorism.Region).plot(kind = "area", stacked=False, figsize=(16,12))
plt.title("Region wise Attack")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
terrorism[["Year","Wound"]].groupby("Year").sum().plot(kind = "bar", color = "grey", figsize=(16,6))
plt.title("Year wise people wounded")
plt.xlabel("Year")
plt.ylabel("Number of people wounded")
plt.show()
```

```python
terrorism[["Year","Killed"]].groupby("Year").sum().plot(kind = "bar", color = "green", figsize=(16,6))
plt.title("Year wise Kills")
plt.xlabel("Year")
plt.ylabel("Number of people kills")
plt.show()
```
