```python
#creating new column this will be helpful when creating visuals in slide
def hour_cat(hour):
    if hour >= 0 and hour <= 7:
        return 'Midnight-Morning'
    elif hour >= 8 and hour <= 13:
        return 'First Half'
    elif hour >= 14 and hour <=17:
        return 'Second Half'
    else:
        return 'Night'
    
df['hour_cat']= df['hour'].apply(hour_cat) #creating a new column applying changes to it
df[['hour','hour_cat']].head(10)
```

```python
#create a unique order id concatenating timestmap, user id and ip address separated by -
df['order_id']= df[['timestamp','user ID','ip_address']].astype(str).agg('-'.join, axis=1)
df.head(10)
```

```python
#create a new column for coupon tag and weekend yes or No

df['coupon_tag']= np.where(df['Order_Coupon_Code'].notna(), 'Yes', 'No')
df['weekend']=np.where(df['day_name'].isin(['Saturday', 'Sunday']), 'Weekend', 'Weekday')
df.head(10)
```

```python
# creating a quantity column to check no of products ordered in a txn
def sum_numbers(product_str):
    # Split the string by comma, convert to integers, and sum
    return sum(int(x) for x in product_str.split(','))

# Apply the function to create the 'quantity' column
df['quantity'] = df['Number_of_Products'].apply(sum_numbers)

# Display the DataFrame
display(df)
```

```python
#check how many users are doing multi txn

user_counts = df['user ID'].value_counts()
# Step 2: Map the counts back to the original DataFrame
df['user_count'] = df['user ID'].map(user_counts)

# Step 3: Use np.where to create the 'multi_txn_user' column
df['multi_txn_user'] = np.where(df['user_count'] > 1, "Yes", "No")

# Drop the temporary 'user_count' column if you don't need it
df.drop(columns=['user_count'], inplace=True)

display(df)
```

## Product Affinity - Apriori Prep

```python
# For Apriori we need to convert to list format, which is necessary for the apriori() function.

df_c=pd.concat([df["order_id"], df["Number_of_Products"].str.split(",", expand = True)], axis=1)
df_c.head(10)
```

```python
df_f=df_c.melt(id_vars=["order_id"], 
        var_name="Product_Split", 
        value_name="Number_of_Products")
df_f.head(25)
```

```python
df_c=pd.concat([df["order_id"], df["Product_Name"].str.split(",", expand = True)], axis=1)
df_c.head(10)
```

```python
df_e=df_c.melt(id_vars=["order_id"], 
        var_name="Product_Split", 
        value_name="Product_Name")
display(df_e.head(10))
display(df_e[df_e.order_id=='2017-10-21 21:08:13-0-49.180.132.205'])
```

```python
df_d=pd.concat([df["order_id"], df["Product_ID"].str.split(",", expand = True)], axis=1)
display(df_d.head(10))
#display(df_d[df_d.order_id=='2017-10-21 21:08:13-0-49.180.132.205'])
```

```python
df_g=df_d.melt(id_vars=["order_id"], 
        var_name="Product_Split", 
        value_name="Product_ID")
display(df_g.head(10))
#display(df_g[df_g.order_id=='2017-10-21 21:08:13-0-49.180.132.205'])
```

```python
a1 = pd.merge(df_f,df_e,on=['order_id',"Product_Split"],how='left')
a2 = pd.merge(a1, df_g,on=['order_id',"Product_Split"],how='left')
a2["Number_of_Products"] =  pd.to_numeric(a2["Number_of_Products"], errors='coerce')
a2["Number_of_Products"] = a2["Number_of_Products"].replace(np.nan, 0)
display(sum(a2["Number_of_Products"] ))
display(a2.head(10))
display(a2.shape)
#display(a2[a2.order_id=='2017-10-21 21:08:13-0-49.180.132.205'])
```

```python
Quantity_to_join = a2.groupby(["order_id"]).agg({"Number_of_Products":sum}).reset_index()

Quantity_to_join.rename({"Number_of_Products" : "Quantity"}, axis='columns', inplace = True)

df_new = pd.merge(df,Quantity_to_join,on=['order_id'],how='left')
display(df_new.head(10))
#display(df_new[df_new.order_id=='2017-10-21 21:08:13-0-49.180.132.205'])
```

## Segmentation

```python
#These are the various segments I will propose and their target customer audience. This is the logic for segments.
#1. Multi txn users 2. coupon users 3. Early birds 4.Midnight shoppers 5.Family audience 6. outlier -shoppers with 0 revenue

df['segment_tag'] = np.where(df.Revenue == 0, "Outlier", 
                    np.where(df.multi_txn_user == "Yes", "Multi_Txns", 
                    np.where(df.coupon_tag == "Yes", "Coupon_Cust",
                    np.where(df.hour_cat == "First Half", "First Half",
                    np.where(df.hour_cat== "Midnight-Morning", "Mid - Night - Morning",
                    np.where((df.hour_cat == "Night") | (df.hour_cat=="Second Half"), "Second Half", "Outlier")
                            )
                            )
                            )
                            )
                            )
df.head((10))
```
